```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi)
sin_ = np.sin(x)
cos_ = np.cos(x)
tan_ = np.tan(x)
plt.plot(x, sin_, x, cos_, x, tan_)
plt.legend(['sin x', 'cos x', 'tan x'])
# Other way
#plt.plot(x, sin_, label = 'sin x')
#plt.plot(x, cos_, label = 'cos x')
#plt.plot(x, tan_, label = 'tan x')
#plt.legend()
```
